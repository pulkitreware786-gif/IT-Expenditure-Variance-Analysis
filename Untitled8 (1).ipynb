```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data from the provided Excel file
df = pd.read_excel('/content/IT Expenditure dataset.xlsx')

print("Data loaded successfully from 'IT Expenditure dataset.xlsx'.")
display(df.head())
```

```python
print("\n--- DataFrame Info ---")
df.info()

print("\n--- Missing Values ---")
print(df.isnull().sum())

print("\n--- Descriptive Statistics ---")
display(df.describe(include='all'))
```

```python
# Convert 'Date' column to a numerical month (1-12)
month_mapping = {
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
    'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
}
df['Month'] = df['Date'].map(month_mapping)

# Fill NaN values in 'Actual', 'Forecast', and 'Plan' columns with 0
financial_cols = ['Actual', 'Forecast', 'Plan']
for col in financial_cols:
    df[col] = df[col].fillna(0)

print("\n--- After Data Cleaning ---")
df.info()
display(df.head())
```

```python
# Calculate variances
df['Actual_vs_Plan'] = df['Actual'] - df['Plan']
df['Actual_vs_Forecast'] = df['Actual'] - df['Forecast']

print("Variances calculated successfully. Displaying the first few rows with new columns:")
display(df.head())
```

## 1. Analysis of Variances by Month

```python
monthly_variances = df.groupby('Month')[['Actual_vs_Plan', 'Actual_vs_Forecast']].sum().reset_index()

print("Monthly Variances (Actual vs Plan and Actual vs Forecast):")
display(monthly_variances)

plt.figure(figsize=(12, 6))
plt.plot(monthly_variances['Month'], monthly_variances['Actual_vs_Plan'], marker='o', label='Actual vs Plan Variance')
plt.plot(monthly_variances['Month'], monthly_variances['Actual_vs_Forecast'], marker='x', label='Actual vs Forecast Variance')
plt.title('Monthly IT Expenditure Variances')
plt.xlabel('Month')
plt.ylabel('Variance Amount')
plt.xticks(monthly_variances['Month'], ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## 2. Analysis of Variances by Business Area

```python
business_area_variances = df.groupby('Business Area')[['Actual_vs_Plan', 'Actual_vs_Forecast']].sum().reset_index()

print("Variances by Business Area:")
display(business_area_variances.sort_values(by='Actual_vs_Plan', ascending=False))

plt.figure(figsize=(14, 7))
sns.barplot(x='Business Area', y='Actual_vs_Plan', data=business_area_variances.sort_values(by='Actual_vs_Plan', ascending=False), palette='viridis')
plt.title('IT Expenditure Variance (Actual vs Plan) by Business Area')
plt.xlabel('Business Area')
plt.ylabel('Variance Amount (Actual vs Plan)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 7))
sns.barplot(x='Business Area', y='Actual_vs_Forecast', data=business_area_variances.sort_values(by='Actual_vs_Forecast', ascending=False), palette='magma')
plt.title('IT Expenditure Variance (Actual vs Forecast) by Business Area')
plt.xlabel('Business Area')
plt.ylabel('Variance Amount (Actual vs Forecast)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## 3. Analysis of Variances by Country

```python
country_variances = df.groupby('Country')[['Actual_vs_Plan', 'Actual_vs_Forecast']].sum().reset_index()

print("Variances by Country:")
display(country_variances.sort_values(by='Actual_vs_Plan', ascending=False))

plt.figure(figsize=(14, 7))
sns.barplot(x='Country', y='Actual_vs_Plan', data=country_variances.sort_values(by='Actual_vs_Plan', ascending=False), palette='viridis')
plt.title('IT Expenditure Variance (Actual vs Plan) by Country')
plt.xlabel('Country')
plt.ylabel('Variance Amount (Actual vs Plan)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 7))
sns.barplot(x='Country', y='Actual_vs_Forecast', data=country_variances.sort_values(by='Actual_vs_Forecast', ascending=False), palette='magma')
plt.title('IT Expenditure Variance (Actual vs Forecast) by Country')
plt.xlabel('Country')
plt.ylabel('Variance Amount (Actual vs Forecast)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## 4. Analysis of Variances by Cost Element

```python
cost_element_variances = df.groupby('Cost element name')[['Actual_vs_Plan', 'Actual_vs_Forecast']].sum().reset_index()

print("Variances by Cost Element:")
display(cost_element_variances.sort_values(by='Actual_vs_Plan', ascending=False).head(10))

plt.figure(figsize=(16, 8))
sns.barplot(x='Cost element name', y='Actual_vs_Plan', data=cost_element_variances.sort_values(by='Actual_vs_Plan', ascending=False).head(10), palette='viridis')
plt.title('Top 10 IT Expenditure Variance (Actual vs Plan) by Cost Element')
plt.xlabel('Cost Element Name')
plt.ylabel('Variance Amount (Actual vs Plan)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

plt.figure(figsize=(16, 8))
sns.barplot(x='Cost element name', y='Actual_vs_Forecast', data=cost_element_variances.sort_values(by='Actual_vs_Forecast', ascending=False).head(10), palette='magma')
plt.title('Top 10 IT Expenditure Variance (Actual vs Forecast) by Cost Element')
plt.xlabel('Cost Element Name')
plt.ylabel('Variance Amount (Actual vs Forecast)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
